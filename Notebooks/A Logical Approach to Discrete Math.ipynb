## A Logical Approach to Discrete Math Ch.3

The following proofs are examples and exercises from chapter 3 of the textbook *A Logical Approach To Discrete Math* by David Gries and Fred B.Schneider.

```fsharp
open Sylvester
open PropCalculus

// Declare symbols we can reuse in formulae
let p,q,r = var3<bool>
```

```fsharp
// Theorem 3.2
let ``3.2`` = proof prop_calculus <@ p = p = q = q @>  [
    RightAssoc |> LR
    def_true <@ q @> |> Trn |> R
    Commute |> LR
]
```

```fsharp
// Theorem 3.4
let ``3.4`` = proof prop_calculus <@ true @> [
    def_true <@ p @> |> LR
]
```

```fsharp
// Theorem 3.5
let ``3.5`` = axiom prop_calculus <@ p = p @>  
```

```fsharp
// Theorem 3.11
let ``3.11`` = proof prop_calculus <@ not p = q = p = not q @> [
        Collect |> L
        RightAssoc |> LR
        Commute |> R
        Collect |> R
        Commute |> R
    ] 
```

```fsharp
// Theorem 3.12 
let ``3.12`` = proof prop_calculus <@ not (not p) = p @> [
        Collect |> LR
        def_false <@ p @> |> Trn |> LR
        not_false |> Truth |> LR
    ]
```

```fsharp
// Theorem 3.13
let ``3.13``=  proof prop_calculus <@not false = true@> [
        Commute |> LR
        def_true <@ false @> |> L
        RightAssoc |> LR
        Commute |> R
        Collect |> R
        def_true <@ false @> |> Trn |> R 
]
```

```fsharp
// Theorem 3.14
let ``3.14`` = axiom prop_calculus <@ (p <> q) = not (p = q) @>
```

```fsharp
// Theorem 3.15
let ``3.15`` = proof prop_calculus <@ not p = p = false@>  [
    Collect |> L
    def_true <@ p @> |> Trn |> L
    Commute |> LR
]
```

```fsharp
// Theorem 3.16
let ``3.16`` = proof prop_calculus <@ (p <> q) = (q <> p) @> [
    def_not_eq <@ p @> <@ q @> |> L
    def_not_eq <@ q @> <@ p @> |> R
    Commute |> R
]
```

```fsharp
// Theorem 3.17
let ``3.17`` = 
    proof prop_calculus <@ ((p <> q) <> r) = (p <> (q <> r)) @> [
        def_not_eq <@ p @> <@ q @> |> L
        def_not_eq <@ not (p = q) @> <@ r @> |> L
        def_not_eq <@ q @> <@ r @> |> R
        def_not_eq <@ p @> <@ not (q = r) @> |> R
        distrib_not <@ q @> <@ r @> |> R
        LeftAssoc |> R
        commute_eq <@ p @> <@ not q @> |> R
        collect_not <@ q @> <@ p @> |> R
        commute_eq <@ q @> <@ p @> |> R
    ]
```

```fsharp
// Theorem 3.29
let ``3.29`` = proof prop_calculus <@ p ||| true = true @> [
    def_true <@ p @> |> LR 
    Distrib |> LR
    Idemp |> LR
]
```

```fsharp
// Theorem 3.30
let ``3.30`` = proof prop_calculus <@p ||| false = p @> [
        def_false <@ p @> |> L
        L Distrib
        LR RightAssoc
        R Idemp
        L ExcludedMiddle
    ]
```

```fsharp
// Theorem 3.31
let ``3.31`` = proof prop_calculus <@ (p ||| (q ||| r)) = ((p ||| q) ||| (p ||| r)) @> [
    commute_or_or <@ p @> <@ q @> <@ p @> <@ r @> |> R
    idemp_or <@ p @> |> R
]
```

```fsharp
let ``3.32`` = proof prop_calculus <@ (p ||| q) = (p ||| not q = p) @> [
    LeftAssoc |> LR
    collect_or_eq <@ p @> <@ q @> <@ not q @> |> LR
    commute_eq <@ q @> <@ not q @> |> L
    def_false <@ q @> |> Trn |> L
    ident_or <@ p @> |> L
]
```

```fsharp
let ``3.37`` = proof prop_calculus <@ (p |&| q |&| r) = (p |&| (q |&| r)) @> [
    golden_rule <@ p @> <@ q @> |> L
    golden_rule <@ (p = q) = (p ||| q) @> <@ r @> |> L 
    commute_or <@ ((p = q) = (p ||| q)) @> <@ r @> |> L
    distrib_or_eq <@ r @> <@ p = q @> <@ p ||| q @> |> L
    distrib_or_eq <@ r @> <@ p @> <@ q @> |> L
    right_assoc_eq <@ p = q@> <@ p ||| q@> <@ r @> |> L
    commute_eq <@ p ||| q @> <@ r @> |> L
    commute_or <@ r @> <@ q @> |> L
    commute_eq <@ r ||| p@> <@ q ||| r@> |> L
    commute_or <@ r @> <@ p ||| q @> |> L
    L LeftAssoc
    L LeftAssoc
    L LeftAssoc
    commute_and <@ p @> <@ q |&| r @> |> R
    ident_and_eq_all <@ q @> <@ r @> <@ p @> |> R
    commute_eq <@q = r@> <@ p @> |> R
]
```

```fsharp
let ``3.38`` = proof prop_calculus <@ (p |&| p) = p @> [
    LR GoldenRule
    RightAssoc |> LR
    idemp_or <@ p @> |> Taut |> R
    LR Commute
]
```

```fsharp
let ``3.39`` = proof prop_calculus <@ (p |&| true) = p @> [
    LR GoldenRule
    LR RightAssoc
    zero_or <@ p @> |> R
    R Commute
]
```

```fsharp
let ``3.40`` = proof prop_calculus <@ (p |&| false) = false @> [
  LR GoldenRule
  ident_or <@ p @> |> L
  LR RightAssoc
]
```

```fsharp
// Theorem 3.41
let ``3.41`` = proof prop_calculus <@ (p |&| (q |&| r)) = ((p |&| q) |&| (p |&| r)) @> [
        LeftAssoc |> R
        right_assoc_and <@ p @> <@ q @> <@ p @> |> R
        commute_and <@ q @> <@ p @> |> R
        LeftAssoc |> R
        idemp_and <@ p @> |> R
        LeftAssoc |> L
]
```

```fsharp
let ``3.42`` = proof prop_calculus <@p |&| not p = false@> [
    GoldenRule |> L
    ExcludedMiddle |> L
    commute_eq <@ p @> <@ not p @> |> L
    def_false <@ p @> |> Trn |> L
    commute_eq <@ false @> <@ true @> |> L
    RightAssoc |> LR
]
```

```fsharp
let ``3.43a`` = proof prop_calculus <@ (p |&| (p ||| q)) = p @> [
    GoldenRule |> L
    left_assoc_or <@ p @> <@ p @> <@ q @> |> L
    idemp_or <@ p @> |> L
]
```

```fsharp
let ``3.43b`` = proof prop_calculus <@ (p ||| (p |&| q)) = p @> [
    GoldenRule |> L
    Distrib |> L
    left_assoc_or <@ p @> <@ p @> <@ q @> |> L
    idemp_or <@ p @> |> L
    Distrib |> L
    idemp_or <@ p @> |> L
]
```
